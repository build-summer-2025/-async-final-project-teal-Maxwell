```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset. Replace 'titanic.csv' with your own dataset file for your project.
# Ensure the dataset file exists and is in the same directory as this jupyter notebook.
car_df = pd.read_csv('car_dataset_2000_rows.csv')


# Show all rows in the dataframe
pd.set_option('display.max_rows', None)
display(car_df)

car_df['price_dollars'] = car_df['price'] / 100


###QUESTION 1
average_price_per_brand = car_df.groupby("brand")["price_dollars"].mean().sort_values(ascending=False)


display(average_price_per_brand)


plt.plot(average_price_per_brand.index, average_price_per_brand.values, color='blue', marker='o')
plt.title('Average Price Per Brand')
plt.xlabel('Brand Name')
plt.ylabel('Average Price (Thousands)')
plt.xticks(rotation=30)
plt.grid(True)
plt.show()


###QUESTION 2
average_price_per_car_type = car_df.groupby("car_type")["price_dollars"].mean().sort_values(ascending=False)


plt.plot(average_price_per_car_type.index, average_price_per_car_type.values, color='blue', marker='o')
plt.title('Average Price Per Car Type')
plt.xlabel('Car Type')
plt.ylabel('Average Price (Thousands)')
plt.xticks(rotation=30)
plt.grid(True)
plt.show()

#####QUESTION 3
average_enginze_size_per_car_type = car_df.groupby("car_type")["engine_size"].mean().sort_values(ascending=False)

car_type = average_enginze_size_per_car_type.index 
engine_size = average_enginze_size_per_car_type.values


fix, ax = plt.subplots()

bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:purple']
label = car_type
ax.set_ylabel('Average Engine Size (Liters)')
ax.set_xlabel('Car Type')  
ax.set_title('Average Engine Size by Car Type')
ax.bar(car_type, engine_size, label=label, color=bar_colors)



plt.xticks(rotation=30)
plt.show()

###QUESTION 4

average_price_each_year = car_df.groupby("year")["price_dollars"].mean().reset_index()



plt.plot(average_price_each_year["year"], average_price_each_year["price_dollars"], color="orange",) 

plt.xlabel("Car Year")
plt.ylabel("Price (Thousands)")
plt.title("Relationship of the Year and Price")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show


###QUESTION 5

plt.figure(figsize=(8, 5))

average_price_per_transmission = (car_df.groupby("transmission")["price_dollars"].mean().sort_values(ascending=False).reset_index())


plt.barh(average_price_per_transmission["transmission"],average_price_per_transmission["price_dollars"], color="red")
plt.title('Average Price Per Transmission')  
plt.xlabel('Transmission Type')
plt.ylabel('Average Price (Thousands)')
plt.xticks(rotation=30)
plt.yticks(rotation=0)
plt.grid(True)
plt.show()

####QUESTION 6



```
